```python
# ==============================================================================
# 0. KAGGLE ENVIRONMENT SETUP: INSTALL LIBRARIES
# ==============================================================================
!pip install torchxrayvision -q

# ==============================================================================
# 1. IMPORTS AND SETUP
# ==============================================================================
import torch
from torch import nn
from torch.utils.data import Dataset, DataLoader
from torch.optim import AdamW

import pandas as pd
import numpy as np
import cv2
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import (roc_auc_score, roc_curve, classification_report,
                             confusion_matrix, precision_recall_curve, average_precision_score)
from tqdm import tqdm
import os
import glob
import timm

# The library for pre-trained x-ray models
import torchxrayvision as xrv

from einops import rearrange
from einops.layers.torch import Rearrange
import albumentations as A
from albumentations.pytorch import ToTensorV2

sns.set_style("whitegrid")

# ==============================================================================
# 2. DATA HANDLING AND VISUALIZATION
# ==============================================================================

def show_image_grid(df, labels, n_images=25, random_state=42):
    plt.figure(figsize=(15, 15)); sample_df = df.sample(n_images, random_state=random_state)
    for i, (idx, row) in enumerate(sample_df.iterrows()):
        plt.subplot(5, 5, i + 1); image_path = row['FullPath']; image = cv2.imread(image_path)
        if image is None: continue
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        image_labels = [label for label in labels if row[label] == 1]
        title = "\n".join(image_labels) if image_labels else "No Finding"
        plt.imshow(image); plt.title(title, fontsize=8); plt.axis('off')
    plt.tight_layout(); plt.show()

class NIHChestXrayDataset(Dataset):
    def __init__(self, df, labels, transform=None, is_val=False):
        self.df, self.transform, self.labels, self.is_val = df, transform, labels, is_val
    def __len__(self): return len(self.df)
    def __getitem__(self, idx):
        row = self.df.iloc[idx]; image_path = row['FullPath']
        image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE); image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)
        original_image = cv2.resize(image, (CONFIG['IMAGE_SIZE'], CONFIG['IMAGE_SIZE']))
        if self.transform: image = self.transform(image=image)['image']
        labels_vector = torch.tensor(row[self.labels].values.astype(np.float32))
        return (image, labels_vector, original_image) if self.is_val else (image, labels_vector)

# <<<--- THIS PREPROCESSING IS CORRECT for the torchxrayvision backbone ---<<<
def get_transforms(image_size):
    """
    Defines augmentations that scale the image to the [0, 1] float range,
    matching the torchxrayvision model's expectation.
    """
    train_transform = A.Compose([
        A.Resize(image_size, image_size),
        A.HorizontalFlip(p=0.5), A.Rotate(limit=15, p=0.7),
        A.RandomBrightnessContrast(brightness_limit=0.1, contrast_limit=0.1, p=0.7),
        A.ToFloat(max_value=255.0), # Scale to [0, 1]
        ToTensorV2()
    ])
    val_transform = A.Compose([
        A.Resize(image_size, image_size),
        A.ToFloat(max_value=255.0), # Scale to [0, 1]
        ToTensorV2()
    ])
    return train_transform, val_transform

# ==============================================================================
# 3. MODEL ARCHITECTURE: SemaCheX-Former (with NIH Backbone)
# ==============================================================================

class TransformerEncoderBlock(nn.Module):
    def __init__(self, dim, num_heads, mlp_dim, dropout=0.1):
        super().__init__(); self.norm1, self.attn, self.norm2 = nn.LayerNorm(dim), nn.MultiheadAttention(dim, num_heads, dropout=dropout, batch_first=True), nn.LayerNorm(dim)
        self.mlp = nn.Sequential(nn.Linear(dim, mlp_dim), nn.GELU(), nn.Dropout(dropout), nn.Linear(mlp_dim, dim), nn.Dropout(dropout))
    def forward(self, x):
        normed_x = self.norm1(x); x = x + self.attn(normed_x, normed_x, normed_x)[0]; x = x + self.mlp(self.norm2(x)); return x

class PartitionReconstructionAttentionBlock_LMSA(nn.Module):
    def __init__(self, in_channels, dim, patch_size, num_heads, mlp_dim, num_classes, dropout=0.1):
        super().__init__(); self.patch_size, patch_dim = patch_size, in_channels * (patch_size ** 2)
        self.partition = Rearrange('b c (h p1) (w p2) -> b (h w) (p1 p2 c)', p1=patch_size, p2=patch_size)
        self.patch_projection = nn.Linear(patch_dim, dim); self.label_embeddings = nn.Parameter(torch.randn(1, num_classes, dim))
        self.attention_module = TransformerEncoderBlock(dim, num_heads, mlp_dim, dropout)
    def forward(self, x):
        b, _, h, w = x.shape; tokens, label_embs = self.partition(x), self.label_embeddings.expand(b, -1, -1)
        tokens = self.patch_projection(tokens); processed_tokens = self.attention_module(torch.cat((tokens, label_embs), dim=1))
        processed_visual_tokens = processed_tokens[:, :tokens.shape[1]]; recon_h, recon_w = h // self.patch_size, w // self.patch_size
        return rearrange(processed_visual_tokens, 'b (h w) d -> b d h w', h=recon_h, w=recon_w)

class ConvSEBlock(nn.Module):
    def __init__(self, in_channels, out_channels, kernel_size=3, reduction=8):
        super().__init__()
        self.conv_block = nn.Sequential(nn.Conv2d(in_channels, out_channels, kernel_size, padding='same', bias=False), nn.BatchNorm2d(out_channels), nn.ReLU(inplace=True))
        self.se = nn.Sequential(nn.AdaptiveAvgPool2d(1), nn.Conv2d(out_channels, out_channels // reduction, 1, bias=False), nn.ReLU(inplace=True), nn.Conv2d(out_channels // reduction, out_channels, 1, bias=False), nn.Sigmoid())
    def forward(self, x): cnn_out = self.conv_block(x); return cnn_out * self.se(cnn_out)

# <<<--- THIS IS YOUR MODEL, NOW WITH THE NIH-TRAINED BACKBONE ---<<<
class SemaCheXFormer(nn.Module):
    def __init__(self, num_classes=14, dim=256, num_heads=8, mlp_dim=512, cnn_channels=256):
        super().__init__()
        # 1. Load the full NIH-trained model
        pretrained_model = xrv.models.DenseNet(weights="densenet121-res224-nih")
        
        # 2. Steal its feature extractor to be our backbone
        self.backbone = pretrained_model.features
        
        # 3. CRITICAL FIX: The first layer of the stolen backbone must be adapted
        #    from its original 1-channel to our 3-channel input.
        #    We replace BOTH the conv and the batchnorm layer.
        self.backbone.conv0 = nn.Conv2d(3, 64, kernel_size=7, stride=2, padding=3, bias=False)
        self.backbone.norm0 = nn.BatchNorm2d(64) # Must be 64 to match the output of conv0
        
        # The output of DenseNet121's feature block is 1024 channels
        backbone_out_channels = 1024
        
        # 4. Build YOUR custom head
        self.attention_stage = PartitionReconstructionAttentionBlock_LMSA(backbone_out_channels, dim, 1, num_heads, mlp_dim, num_classes)
        self.cnn_se_stage = ConvSEBlock(dim, cnn_channels)
        self.classification_head = nn.Sequential(nn.AdaptiveAvgPool2d(1), nn.Flatten(), nn.Dropout(p=0.2, inplace=True), nn.Linear(cnn_channels, num_classes))

    def forward(self, x):
        # The entire feature extraction is now handled by the expert backbone
        features = self.backbone(x)
        
        # The rest is YOUR architecture
        attention_out = self.attention_stage(features)
        cnn_out = self.cnn_se_stage(attention_out)
        return self.classification_head(cnn_out)

# ==============================================================================
# 4. TRAINING AND EVALUATION LOOPS
# ==============================================================================
def train_one_epoch(model, dataloader, criterion, optimizer, device, scaler):
    model.train(); total_loss, all_preds, all_labels = 0, [], []
    for images, labels in tqdm(dataloader, desc="Training"):
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad(set_to_none=True)
        with torch.cuda.amp.autocast(): outputs = model(images); loss = criterion(outputs, labels)
        scaler.scale(loss).backward(); scaler.step(optimizer); scaler.update()
        total_loss += loss.item(); all_preds.append(torch.sigmoid(outputs).detach().cpu().numpy()); all_labels.append(labels.cpu().numpy())
    avg_loss, all_preds, all_labels = total_loss/len(dataloader), np.vstack(all_preds), np.vstack(all_labels)
    accuracy = ((all_preds > 0.5).astype(int) == all_labels).mean()
    return avg_loss, accuracy
def evaluate(model, dataloader, criterion, device):
    model.eval(); total_loss, all_preds, all_labels, all_original_images = 0, [], [], []
    with torch.no_grad():
        for images, labels, original_images in tqdm(dataloader, desc="Validating"):
            images, labels = images.to(device), labels.to(device)
            with torch.cuda.amp.autocast(): outputs = model(images); loss = criterion(outputs, labels)
            total_loss += loss.item(); all_preds.append(torch.sigmoid(outputs).cpu().numpy()); all_labels.append(labels.cpu().numpy()); all_original_images.append(original_images.numpy())
    avg_loss, all_preds, all_labels, all_original_images = total_loss/len(dataloader), np.vstack(all_preds), np.vstack(all_labels), np.vstack(all_original_images)
    accuracy = ((all_preds > 0.5).astype(int) == all_labels).mean()
    return avg_loss, accuracy, all_preds, all_labels, all_original_images
# ==============================================================================
# 5. PLOTTING AND METRICS REPORTING
# ==============================================================================
def plot_history(history):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6)); ax1.plot(history['train_loss'], 'o--', label='Train Loss'); ax1.plot(history['val_loss'], 'o-', label='Validation Loss')
    ax1.set_title('Model Loss'); ax1.set_xlabel('Epochs'); ax1.legend(); ax2.plot(history['train_acc'], 'o--', label='Train Accuracy'); ax2.plot(history['val_acc'], 'o-', label='Validation Accuracy')
    ax2.set_title('Model Accuracy'); ax2.set_xlabel('Epochs'); ax2.legend(); plt.show()
def plot_roc_curves(y_true, y_pred, labels):
    plt.figure(figsize=(12, 10))
    for i, label in enumerate(labels):
        if len(np.unique(y_true[:, i]))<2: continue
        fpr,tpr,_=roc_curve(y_true[:,i],y_pred[:,i]); plt.plot(fpr,tpr,label=f'{label} (AUC={roc_auc_score(y_true[:,i],y_pred[:,i]):.3f})')
    plt.plot([0,1],[0,1],'k--',label='Chance'); plt.title('ROC Curves'); plt.xlabel('FPR'); plt.ylabel('TPR'); plt.legend(loc='lower right'); plt.show()
def plot_precision_recall_curves(y_true, y_pred, labels):
    plt.figure(figsize=(12, 10))
    for i, label in enumerate(labels):
        if len(np.unique(y_true[:, i]))<2: continue
        prec,rec,_=precision_recall_curve(y_true[:,i],y_pred[:,i]); plt.plot(rec,prec,label=f'{label} (AP={average_precision_score(y_true[:,i],y_pred[:,i]):.3f})')
    plt.title('Precision-Recall Curves'); plt.xlabel('Recall'); plt.ylabel('Precision'); plt.legend(loc='upper right'); plt.show()
def plot_confusion_matrices(y_true, y_pred, labels):
    y_pred_binary=(y_pred>0.5).astype(int); n_labels,n_cols=len(labels),4; n_rows=n_labels//n_cols+(1 if n_labels%n_cols else 0)
    fig,axes=plt.subplots(n_rows,n_cols,figsize=(20,5*n_rows)); axes=axes.flatten()
    for i,label in enumerate(labels):
        cm=confusion_matrix(y_true[:,i],y_pred_binary[:,i]); sns.heatmap(cm,annot=True,fmt='d',ax=axes[i],cmap='Blues')
        axes[i].set_title(label); axes[i].set_xlabel('Predicted'); axes[i].set_ylabel('True')
    for j in range(i+1,len(axes)): fig.delaxes(axes[j])
    plt.tight_layout(); plt.show()
def print_and_plot_metrics(y_true, y_pred, labels):
    y_pred_binary=(y_pred>0.5).astype(int); print("Classification Report\n"+"="*70); print(classification_report(y_true,y_pred_binary,target_names=labels,zero_division=0))
    auc_scores=[roc_auc_score(y_true[:,i],y_pred[:,i]) if len(np.unique(y_true[:,i]))>1 else np.nan for i in range(len(labels))]; mean_auc=np.nanmean(auc_scores)
    print("\nROC AUC Scores\n"+"="*70)
    for i,label in enumerate(labels): print(f"{label:<20}: {auc_scores[i]:.4f}" if not np.isnan(auc_scores[i]) else f"{label:<20}: Not Defined")
    print(f"\n{'MEAN AUC':<20}: {mean_auc:.4f}\n")
    report_df=pd.DataFrame(classification_report(y_true,y_pred_binary,target_names=labels,output_dict=True,zero_division=0)).iloc[:-1,:].T
    plt.figure(figsize=(10,8)); sns.heatmap(report_df,annot=True,cmap='viridis',fmt='.2f'); plt.title('Metrics Heatmap'); plt.show()

# ==============================================================================
# 6. MAIN EXECUTION SCRIPT
# ==============================================================================
if __name__ == '__main__':
    
    training_successful = False
    
    try:
        CONFIG = {
            "IMAGE_SIZE": 224, "BATCH_SIZE": 24, "LEARNING_RATE": 0.001,
            "EPOCHS": 10,
            "DEVICE": "cuda" if torch.cuda.is_available() else "cpu",
            "DATASET_DIR": "/kaggle/input/data",
            "CSV_PATH": "/kaggle/input/data/Data_Entry_2017.csv",
        }

        print(f"--- Training SemaCheXFormer with NIH Pre-trained Backbone | Device: {CONFIG['DEVICE']} ---")
        
        try: df = pd.read_csv(CONFIG['CSV_PATH'])
        except FileNotFoundError:
            print("="*80 + "\nFATAL ERROR: Could not find CSV. Please update 'CSV_PATH' and 'DATASET_DIR' in CONFIG.\n" + "="*80); raise

        all_image_paths = {os.path.basename(x): x for x in glob.glob(os.path.join(CONFIG['DATASET_DIR'], 'images*', '*', '*.png'))}
        if not all_image_paths: raise FileNotFoundError("FATAL ERROR: No images found.")

        df['FullPath'] = df['Image Index'].map(all_image_paths.get); df.dropna(subset=['FullPath'], inplace=True)
        all_labels_str = df['Finding Labels'].str.get_dummies(sep='|'); LABELS = [l for l in all_labels_str.columns if l != 'No Finding']
        df = pd.concat([df, all_labels_str[LABELS]], axis=1)
        print(f"Successfully processed {len(df)} images.")
        
        train_df, val_df = df.sample(frac=0.8, random_state=42), df.drop(df.sample(frac=0.8, random_state=42).index)
        print(f"Training size: {len(train_df)} | Validation size: {len(val_df)}")

        train_transform, val_transform = get_transforms(CONFIG["IMAGE_SIZE"])
        train_dataset = NIHChestXrayDataset(train_df, LABELS, transform=train_transform)
        val_dataset = NIHChestXrayDataset(val_df, LABELS, transform=val_transform, is_val=True)
        train_loader = DataLoader(train_dataset, batch_size=CONFIG['BATCH_SIZE'], shuffle=True, num_workers=2, pin_memory=True)
        val_loader = DataLoader(val_dataset, batch_size=CONFIG['BATCH_SIZE'], shuffle=False, num_workers=2, pin_memory=True)

        model = SemaCheXFormer(num_classes=len(LABELS)).to(CONFIG['DEVICE'])
        
        pos_counts = train_df[LABELS].sum(); pos_weights = (len(train_df) - pos_counts) / pos_counts
        criterion = nn.BCEWithLogitsLoss(pos_weight=torch.tensor(pos_weights.values, dtype=torch.float32).to(CONFIG['DEVICE']))
        optimizer = AdamW([
            {'params': model.backbone.parameters(), 'lr': CONFIG['LEARNING_RATE'] / 10},
            {'params': model.attention_stage.parameters(), 'lr': CONFIG['LEARNING_RATE']},
            {'params': model.cnn_se_stage.parameters(), 'lr': CONFIG['LEARNING_RATE']},
            {'params': model.classification_head.parameters(), 'lr': CONFIG['LEARNING_RATE']}])
        scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.1, patience=2, verbose=True)
        scaler = torch.cuda.amp.GradScaler()

        history = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}
        best_val_loss = float('inf')
        print(f"--- Starting training for {CONFIG['EPOCHS']} epochs with AMP ---")
        
        for epoch in range(CONFIG['EPOCHS']):
            print(f"\n--- Epoch {epoch+1}/{CONFIG['EPOCHS']} ---")
            train_loss, train_acc = train_one_epoch(model, train_loader, criterion, optimizer, CONFIG['DEVICE'], scaler)
            val_loss, val_acc, y_pred, y_true, val_images = evaluate(model, val_loader, criterion, CONFIG['DEVICE'])
            
            history['train_loss'].append(train_loss); history['val_loss'].append(val_loss)
            history['train_acc'].append(train_acc); history['val_acc'].append(val_acc)
            print(f"Epoch {epoch+1} | Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f}")
            
            if val_loss < best_val_loss:
                best_val_loss = val_loss
                torch.save(model.state_dict(), 'best_model.pth')
                print(f"ðŸ† Best model saved with val_loss: {best_val_loss:.4f}")
            scheduler.step(val_loss)

        torch.save(model.state_dict(), f'final_model_{CONFIG["EPOCHS"]}_epochs.pth')
        training_successful = True

    finally:
        if training_successful:
            print("\n\n--- ðŸ“ˆ Final Evaluation & Visualization ---")
            model_path = 'best_model.pth'
            if not os.path.exists(model_path): model_path = f'final_model_{CONFIG["EPOCHS"]}_epochs.pth'
            
            if os.path.exists(model_path):
                model = SemaCheXFormer(num_classes=len(LABELS)).to(CONFIG['DEVICE'])
                model.load_state_dict(torch.load(model_path))
                print(f"--- Loaded model from {model_path} for evaluation. ---")
                
                _, _, y_pred, y_true, _ = evaluate(model, val_loader, criterion, CONFIG['DEVICE'])
                
                plot_history(history)
                plot_roc_curves(y_true, y_pred, LABELS)
                plot_precision_recall_curves(y_true, y_pred, LABELS)
                plot_confusion_matrices(y_true, y_pred, LABELS)
                print_and_plot_metrics(y_true, y_pred, LABELS)
            else:
                print("No models were saved. Cannot run final evaluation.")
```
